# Part 1: Forecasting Sales Price Using Seasonal ARIMA Model

```r
library(dplyr)
library(astsa)
library(Cairo)
```

First Step is to load the data, clean/filter, and convert to a time series object. This first dataset I am using is from Redfin and contains median house sales price data from various cities and metro areas.

```r
medHousePrice <- read.csv("sea_housing.csv", fileEncoding = "UTF-8-BOM")
h <- medHousePrice %>%
        filter(Region == "Seattle, WA metro area") %>%
        select(Month.of.Period.End, Median.Sale.Price)
t <- as.numeric(sub('\\$', '', sub('K','',as.character(h$Median.Sale.Price))))*1000
t <- ts(t,start=c(2012,1),end=c(2017,4),frequency=12)
```

Next I plotted the time series to get a feel for the data.

```r
#set option to reduce plot size for better viewing
options(repr.plot.width = 5, repr.plot.height = 4)
```

```r
plot(t, main = "Seattle Median House Price Since 2012", ylab = "Median Sales Price")
```

Ok, right away I notice a few things about this time series:
* There is a clear upward trend
* There is clear seasonality
* variance is approximately constant throughout time series

Given these findings, I will need to transform the time series so that it is stationary.

```r
options(repr.plot.width = 6, repr.plot.height = 6)
d_t <- diff(t)
dd_t <- diff(t, lag = 12)
par(mfrow=c(2,1))
plot(d_t)
plot(dd_t)
```

Ok looks stationary after taking the diff and diff lag-12 out. Now I can look at ACF and PACF to determine which ARIMA model to select.

```r
acf2(dd_t, max.lag = 48)
```

For the non-seasonal component, I would say the ACF and PACF both trail off, so an ARIMA (1,1,1) could work.
For the seasonal component, it looks like the ACF is trailing off and the PACF cuts off after 1, so an AR seasonal component would most likely be appropriate. Now I'll try a (1,1,1) x (1,1,0)12 sarima fit

```r
m1 <- sarima(t, 1, 1, 1, 1, 1, 0, 12)
```

Not bad. The residuals look like white noise for the most part. ACF of residuals is not showing any additional autocorrelation. The test for normality looks ok. Lastly, none of the p-values for the Ljung-Box test are less than 0.05, so we do not reject the null hypothesis that the residuals are independently distributed.

Now I'll try some other models to minimize AIC and BIC.

```r
m2 <- sarima(t, 1, 1, 0, 0, 1, 1, 12)
```

The best model I could find based on AIC and BIC was the (1,1,0) x (0,1,1)12 model. 

The forecast library has an auto.arima function that can sometimes work well. I'll run it up against the best model that I have found.

```r
library(forecast)
auto.arima(t)
```

```r
m3 <- sarima(t,1,1,0,1,0,0,12)
```

```r
sprintf("Model 1: AIC is %f, BIC is %f", m1$AIC, m1$BIC)
sprintf("Model 2: AIC is %f, BIC is %f", m2$AIC, m2$BIC)
sprintf("Model 3 (auto arima): AIC is %f, BIC is %f", m3$AIC, m3$BIC)
```

```r
(544648/506717)
```

```r
f1 <- sarima.for(t,1,1,0,0,1,1,12, n.ahead=12)
```

```r
appreciationForecast <- abs(f1$pred[12]/f1$pred[1]-1)*100
appreciationForecastLow <- abs((f1$pred[12]-2*f1$se[12])/f1$pred[1] - 1)*100
appreciationForecastHigh <- abs((f1$pred[12]+2*f1$se[12])/f1$pred[1] - 1)*100
sprintf("From this model, we estimate house prices to appreciate by %.02f%% next year,
         being 95%% confident that the appreciation rate will be between %.02f%% and %.02f%%",
        appreciationForecast, appreciationForecastLow, appreciationForecastHigh)
```

<b>Conclusion</b>: Throughout this project, I was able to achieve a reasonably good fit with a seasonal ARIMA model. I was able to forecast out 1 year with standard error of about 3% of the prediction. The 95% confidence interval is fairly wide, but good enough for our purposes.

# Part 2: Quantifying Seasonal Difference in Sales Price

```r
par(mfrow=c(2,1))
options(repr.plot.width = 5, repr.plot.height = 6)
trend_t <- ma(t, order = 12, centre = T)
plot(as.ts(t))
lines(trend_t, col = "red")
plot(as.ts(trend_t), col = "red")
```

```r
par(mfrow=c(1,1))
options(repr.plot.width = 3, repr.plot.height = 3)
detrend_t <- t - trend_t
plot(as.ts(detrend_t))
length(detrend_t)
```

```r
m_t = t(matrix(data = detrend_t[5:64], nrow = 12))
seasonal_t = colMeans(m_t, na.rm = T)
plot(as.ts(rep(seasonal_t,5)))
sprintf("Average seasonal difference in median sales price is $%.02f.", max(seasonal_t) - min(seasonal_t))
```

<b>Part 2 Conclusion</b>: After detrending the time series, I was able to find that seasonal difference in median sales price is about $40,000 on average, which could mean a difference of 10% of the price when deciding to sell in the summer peak season versus the winter low season.
